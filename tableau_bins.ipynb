```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
#from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
#from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
#import tensorflow as tf
```

```python
tableau_df = pd.read_csv('combo_df.csv')
tableau_df.head()
```

```python
tableau_df.describe()
```

```python
gymscapitabins = [0, 3, 5, 6, 10]
labels = [1,2,3,4]
tableau_df['gyms_per_capita_bin'] = pd.cut(tableau_df['Gyms per Capita (100,000)'], gymscapitabins, labels=labels)

restaurantpercapitabins = [0, 2.7, 3.4, 5.1, 13]
tableau_df['restaurants_per_capita_bin'] = pd.cut(tableau_df['Restaurants per Capita 100k'], restaurantpercapitabins, labels=labels)

obesitypercapitabins = [0, 29300, 32100, 34800, 40800]
tableau_df['obesity_per_capita_bin'] = pd.cut(tableau_df['obesity per capita 100k'], obesitypercapitabins, labels=labels)


tableau_df.head()
```
