```python
# Downloading the csv file from your GitHub account

url = "https://raw.githubusercontent.com/kelliwolf17/Final_Project/main/Resources/merged_df.csv" # Make sure the url is the raw version of the file on GitHub
download = requests.get(url).content

# Reading the downloaded content and turning it into a pandas dataframe
#data=pd.read_csv("File_path", sep='\t')

df = pd.read_csv(io.StringIO(download.decode('utf-8')), low_memory=False)

# Printing out the first 5 rows of the dataframe

df.head()
```
